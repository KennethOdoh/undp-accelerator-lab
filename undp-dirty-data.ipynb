```python
# EDITED
# Replaced x with other values (instead of 1) in below columns instead of 0. This will maake it more interpretable by others

# Columns that will be combined subsequently
energy_data['Open Source'] = energy_data['Open Source'].replace('x', 'Open Source')
energy_data['IP Protected'] = energy_data['IP Protected'].replace('x', 'IP Protected')
energy_data['Prototype'] = energy_data['Prototype'].replace('x', 'Prototype')
energy_data['Product'] = energy_data['Product'].replace('x', 'Product')
```
